```python
#LambdaRank Learning Rate
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn 

seaborn.set()

lambdarank_models = "results/ml_benchmark20m_2021_11_26T10_36_55/stdout"
bce_models = "./results/ml_benchmark20m_2021_11_27T11_36_03/stdout"
```

```python
from analyze_experiment_in_progress import get_data_from_logs
```

```python
data_lambdarank = pd.DataFrame(get_data_from_logs(lambdarank_models, 0))
data_bce = pd.DataFrame(get_data_from_logs(bce_models, 0))
```

```python
data_lambdarank['learning_rate'] = data_lambdarank.model_name.apply(lambda x: float(x.split(':')[-1]))
data_bce['learning_rate'] = data_bce.model_name.apply(lambda x: float(x.split(':')[-1]))

data_lambdarank = data_lambdarank.sort_values('learning_rate')
data_bce = data_bce.sort_values('learning_rate')
```

```python
plt.figure(figsize = (16, 9))
plt.plot(data_lambdarank.learning_rate, data_lambdarank['ndcg@40'], label = 'Lambdarank')
plt.plot(data_bce.learning_rate, data_bce['ndcg@40'], label = 'BCE')

plt.xlabel('learning rate')
plt.ylabel('ndcg@40')
plt.legend()

plt.title("effect of learning rate on the models with fixed training time(4h)")
```
